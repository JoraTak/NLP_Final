# Pulling Election Tweets

```python
from searchtweets import ResultStream, gen_rule_payload, load_credentials
import json
from searchtweets import collect_results

import numpy as np
import pandas as pd
```

Load Credentials

```python
premium_search_args_production = load_credentials("twitter_keys.yaml",
                                          yaml_key="search_tweets_api_production",
                                          env_overwrite=False)
```

List of candidates we are interested in

```python
candidates = ["Kyrsten Sinema", "Martha McSally",
             "Jacky Rosen", "Dean Heller",
             "Bill Nelson", "Rick Scott",
             "Jon Tester", "Matt Rosendale",
             "Josh Hawley", "Claire McCaskill",
             "Marsha Blackburn", "Phil Bredesen",
             "Mike Braun", "Joe Donnelly",
             "Joe Manchin", "Patrick Morrisey"]
```

API call to pull the Tweets

Uncomment the cell below and run it

**Be careful not to run this accidentally as this counts towards alloted caps!!!**

```python
full_set = []
for cand in candidates:
    rule = gen_rule_payload(cand,
                        from_date="2018-09-05",
                        to_date="2018-10-29",
                        results_per_call=100)
    
    tweets = collect_results(rule, max_results=100, result_stream_args=premium_search_args_production)
    full_set.append(tweets)
```

```python
#To print the tweets in a returned set
#[print(tweet.all_text) for tweet in tweets2[0:10]];
```

```python
data = pd.DataFrame(columns=['candidate','tweet','date'])
```

Put all the tweets into 1 DataFrame

```python
# This code is currently inefficient and needs optimization for bigger datasets
for a in range(len(full_set)):
    temp_data = pd.DataFrame(columns=['candidate','tweet','date'])
    for i in range(11):
        for tweet in full_set[a][i-1:i]:
            temp_data.loc[i] = np.array([candidates[a],tweet.all_text, tweet.created_at_datetime])
    data = data.append(temp_data, ignore_index=True)
```
